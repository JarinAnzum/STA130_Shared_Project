```python
# Insert chart of age breakdown. 
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import data (for example, a CSV file)
df = pd.read_csv("https://raw.githubusercontent.com/pointOfive/stat130chat130/refs/heads/main/CP/CSCS_data_anon.csv", low_memory=False)  # Replace with your actual file path

# Step 2: Define age bins (intervals of 10 years)
bins = range(0, 101, 10)  # This will create intervals: (0-9), (10-19), ..., (90-99)

# Step 3: Categorize participants into these bins
df['Age Group'] = pd.cut(df['DEMO_age'], bins, right=False, labels=[f'{i}-{i+9}' for i in bins[:-1]])

# Step 4: Count how many participants fall into each age group
age_group_counts = df['Age Group'].value_counts().sort_index()

# Step 5: Create the bar graph
plt.figure(figsize=(10, 6))  # Optional: specify the size of the figure
plt.bar(age_group_counts.index, age_group_counts.values)

# Step 6: Customize and show the plot
plt.title('Age Distribution of Participants')
plt.xlabel('Age Group')
plt.ylabel('Number of Participants')
plt.xticks(rotation=45)  # Optional: rotate x-axis labels if necessary
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Show the plot
plt.show()
```

```python
# Insert chart of racial identity breakdown. 
import matplotlib.pyplot as plt

# Step 1: Import data (for example, a CSV file)
df = pd.read_csv("https://raw.githubusercontent.com/pointOfive/stat130chat130/refs/heads/main/CP/CSCS_data_anon.csv", low_memory=False)

# Step 2: Count the number of participants in each ethnicity
ethnicity_counts = df['DEMO_ethnicity'].value_counts()

plt.figure(figsize=(8, 8))  # Optional: set the figure size
plt.pie(ethnicity_counts, labels=ethnicity_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Step 4: Customize and show the plot
plt.title('Distribution of Participants by Ethnicity')

# Show the plot
plt.show()
```
